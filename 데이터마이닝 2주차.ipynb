```python
import seaborn as sns

# seaborn에 내장되어 있는 diamonds 데이터셋 사용
diamonds = sns.load_dataset('diamonds')

# 처음 6개 데이터 사용
diamonds.head(6)

# 마지막 10개 데이터 사용
diamonds.tail(10)
```

```python
# 행의 수 출력
print(diamonds.shape[0])

# 열의 수 출력
print(diamonds.shape[1])

# 마지막 6개 데이터 출력
print(diamonds.tail(6))

# 만약 특정 범위의 행을 직접 인ㄷ덱싱하여 출력하고 싶다면 :
print(diamonds.iloc[53935:53940]) # 범위는 53935 이상 53940 미만
```

```python
# x 열의 값이 0인 행을 출력
diamonds[diamonds['x'] == 0]
```

```python
# x 값이 10보다 큰 모든 행을 출력
diamonds[diamonds['x'] > 10]

# new라는 변수에 데이터를 저장
new = diamonds[diamonds['x'] > 10]
```

```python
new.to_csv("newdiamonds.csv", index = False)
# dataframe.to_csv('파일이름.csv', index = True 설정 시 csv 파일 맨 첫 열에 새로운 인덱스 열이 생성)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 생성
cars = [1, 3, 6, 4, 9]
trucks = [2, 5, 4, 5, 12]
suvs = [4, 4, 6, 6, 16]

# 데이터프레임 생성
autos_data = pd.DataFrame({'cars' : cars, 'trucks' : trucks, 'suvs' : suvs},
                          index = ["Mon", "Tue", "Wed", "Thu", "Fri"])
# 인덱스를 수동으로 설정 및 이름 지정

# 생성된 데이터프레임 확인
autos_data
```

```python
# 단일 막대 그래프, matplotlib 사용
plt.bar(autos_data.index, autos_data['cars']) # x축을 인덱스, y축을 'cars' 열의 값을 사용
plt.show()

# 다중 막대 그래프, matplotlib이 아닌 pandas의 내장 플롯(plot)함수를 사용하여 간단하게 표현 가능
autos_data.plot(kind = 'bar', title = "Autos", ylabel = "Total")
plt.show()

# 누적 막대 그래프
autos_data.plot(kind = 'bar', title = "Autos", ylabel = "Total", stacked = True)
```

```python
# 파이 차트
plt.pie(cars, labels = autos_data.index)
plt.show()

plt.pie(cars, labels = autos_data.index, colors = ['red', 'orange', 'yellow', 'green', 'blue'])
plt.show()
```

```python
# 히스토그램
plt.hist(autos_data.values.flatten(), color = 'orange')
plt.show()

# 확률 히스토그램
plt.hist(autos_data.values.flatten(), density = True, color = 'orange')
plt.show()
```

```python
# 박스플롯
plt.boxplot(autos_data.values.flatten())
plt.show()
```
